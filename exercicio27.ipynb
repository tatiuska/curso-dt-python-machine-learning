```python
# relembrando o uso so matplotlib para criar gráficos
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1000, 1000, 1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
# criando vários gráficos ao mesmo tempo
y1 = x**2
y2 = x**3
y3 = x**2
y4 = x**3
plt.subplot(2, 2, 1) #quantas linhas, quantas colunas, qual o quadrante onde esse gráfico vai ficar (2, 2, 1)
plt.plot(x, y1)
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.show()
```

```python
# criando vários gráficos ao mesmo tempo
y1 = x**2
y2 = x**3
y3 = x**2
plt.subplot(2, 2, 1) #quantas linhas, quantas colunas, qual o quadrante onde esse gráfico vai ficar (2, 2, 1)
plt.plot(x, y1)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.show()
```

```python
y1 = x**2
y2 = x**3
y3 = x**2
y4 = x**3

figura = plt.figure(figsize=(10,10))

figura.add_subplot(2,2,1)
plt.plot(x, y1)
figura.add_subplot(1,2,2)
plt.plot(x, y2)
figura.add_subplot(2,2,3)
plt.plot(x, y3)
#figura.add_subplot(2,2,4)
#plt.plot(x, y4)
plt.show()
```

```python
#usando um loop for para gerar os gráficos
y1 = x**2
y2 = x**3
y3 = x**2
y4 = x**3
y = [y1, y2, y3, y4] #lista

figura = plt.figure(figsize=(10,10))
for i in range(4):
    figura.add_subplot(2,2,i+1)
    plt.plot(x, y[i])
plt.show()
```
